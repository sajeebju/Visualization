```python
import numpy as np
import matplotlib.pyplot as plt 
from PIL import Image
import matplotlib.image as mpimg
import cv2
import os
 

fig = plt.figure(figsize=(10, 10))
plt.subplots_adjust(wspace= 0.1, hspace= 0)

# Automobile image
ax1 = fig.add_subplot(2, 3, 1)
ax1.set_title("Automobile")
ax1.axis("off")
automobile_img = Image.open(os.path.join("/content/drive/MyDrive/satellite images/automobile.jpeg"))
ax1.imshow(automobile_img)

#Railway track track image
ax2 = fig.add_subplot(2, 3, 2)
ax2.set_title("Railway track")
ax2.axis("off")
railway_track_img = Image.open(os.path.join("/content/drive/MyDrive/satellite images/railwaytrack.jpeg")) 
ax2.imshow(railway_track_img)

#BaseBall Court Image
ax3 = fig.add_subplot(2, 3, 3)
ax3.set_title("BaseBall Court ")
ax3.axis("off")
baseball_court_img = Image.open(os.path.join("/content/drive/MyDrive/satellite images/baseball.jpeg"))
ax3.imshow(baseball_court_img)

#Amazon Rain Forest Image
ax4 = fig.add_subplot(2, 3, 4)
ax4.set_title("Amazon Rain Forest")
ax4.axis("off")
amazon_img = Image.open(os.path.join("/content/drive/MyDrive/satellite images/amazon.jpeg"))
ax4.imshow(amazon_img)

#Swimming Pool Image
ax5 = fig.add_subplot(2, 3, 5)
ax5.set_title("Swimming Pool")
ax5.axis("off")      
swimming_pool_img = Image.open(os.path.join("/content/drive/MyDrive/satellite images/swimmingpool.jpeg"))
ax5.imshow(swimming_pool_img)

#Meandering River Image
ax6 = fig.add_subplot(2, 3, 6)
ax6.set_title("Meandering River") 
ax6.axis("off") 
meandering_river_image = Image.open(os.path.join("/content/drive/MyDrive/satellite images/meandering.jpeg"))
ax6.imshow(meandering_river_image)


```

```python
# code for displaying multiple images in one figure

#import libraries
import cv2
from matplotlib import pyplot as plt
import os

# create figure
fig = plt.figure(figsize=(10, 7))

# setting values to rows and column variables
rows = 2
columns = 3

# reading images
Image1 = cv2.imread(os.path.join("/content/drive/MyDrive/satellite images/automobile.jpeg"))
Image2 = cv2.imread(os.path.join("/content/drive/MyDrive/satellite images/railwaytrack.jpeg"))
Image3 = cv2.imread(os.path.join("/content/drive/MyDrive/satellite images/baseball.jpeg"))
Image4 = cv2.imread(os.path.join("/content/drive/MyDrive/satellite images/amazon.jpeg"))
Image5 = cv2.imread(os.path.join("/content/drive/MyDrive/satellite images/swimmingpool.jpeg"))
Image6 = cv2.imread(os.path.join("/content/drive/MyDrive/satellite images/meandering.jpeg"))

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)

# showing image
plt.imshow(Image1)
plt.axis('off')
plt.title("Automobile")

# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)

# showing image
plt.imshow(Image2)
plt.axis('off')
plt.title("Railway Track")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)

# showing image
plt.imshow(Image3)
plt.axis('off')
plt.title("Base Ball Court")

# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)

# showing image
plt.imshow(Image4)
plt.axis('off')
plt.title("Amazon Rain Forest")

# Adds a subplot at the 5th position
fig.add_subplot(rows, columns, 5)

# showing image
plt.imshow(Image5)
plt.axis('off')
plt.title("Swimming Pool")

# Adds a subplot at the 6th position
fig.add_subplot(rows, columns, 6)

# showing image
plt.imshow(Image6)
plt.axis('off')
plt.title("Meandering River")
```
