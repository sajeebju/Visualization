```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#(A) Seismicity to the East (CTFB)
yearA = [2016, 2012, 2012, 2003, 1999, 1955, 1762 ] 

#(B) Seismicity to the North (Surma basin and adjacent region)
yearB = [2011, 1950, 1943, 1934, 1930, 1905, 1897, 1869, 1548, 1505]  

# (C) Seismicity to the West (Foreland Shelf)
yearC = [1935, 1935, 1842, 1842, 1787, 1737] 

# (D) Seismicity at the Centre of the Bengal Basin

yearD = [2008, 1845, 1946, 1845, 1812] 


#Earthquake magnitude of A, B, C, D area respectively

mA = [6.8, 6.8, 6.8, 6.1, 5.2, 6.2, 8.5]
mB = [6.9, 8.4, 7.2, 8.1, 7.1, 7.8, 8.1, 7.5, 0, 8.7]
mC = [6.2, 6.2, 7.3, 7.3, 0, 0 ]
mD = [5.1, 7, 6.2, 7.1, 0]

#DataFrame creating

df1 = pd.DataFrame({"Magnitude" : mA, "Year" : yearA})
df2 = pd.DataFrame({"Magnitude" : mB, "Year" : yearB})
df3 = pd.DataFrame({"Magnitude" : mC, "Year" : yearC})
df4 = pd.DataFrame({"Magnitude" : mD, "Year" : yearD})

#Sorting year values ascending
df1.sort_values("Year", inplace = True)
df2.sort_values("Year", inplace = True)
df3.sort_values("Year", inplace = True)
df4.sort_values("Year", inplace = True)

#Generating subplots

fig, ax = plt.subplots(figsize= (15, 10),nrows=2, ncols=2)
 
df1.plot(x = "Year", y = "Magnitude", ax=ax[0,0], kind='bar', color = 'r', grid=True, rot = 0, title = "Seismicity to the East (CTFB)")
df2.plot(x = "Year", y = "Magnitude", ax=ax[0,1], kind='bar', color = 'c', grid=True, rot = 0,  title =  "Seismicity to the North (Surma basin and adjacent region)")
df3.plot(x = "Year", y = "Magnitude", ax=ax[1,0], kind='bar', color = 'k', grid=True, rot = 0,  title = "Seismicity to the West (Foreland Shelf)")
df4.plot(x = "Year", y = "Magnitude", ax=ax[1,1], kind='bar', color = 'y', grid=True, rot = 0,  title = "Seismicity at the Centre of the Bengal Basin")
```
